```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\acer\\Desktop\\heartdisease.csv")
df.head()
```

```python
plt.subplot(1, 2, 1)
sns.boxplot(df['Age'])
plt.title('Boxplot')

plt.subplot(1, 2, 2)
sns.distplot(df['Age'])
plt.title('Distribution Plot')


# Adjust layout and display plots
plt.tight_layout()
plt.show()
```

##No outliers for Age column

```python
plt.subplot(1, 2, 1)
sns.boxplot(df['RestingBP'])
plt.title('Boxplot')

plt.subplot(1, 2, 2)
sns.distplot(df['RestingBP'])
plt.title('Distribution Plot')


# Adjust layout and display plots
plt.tight_layout()
plt.show()
```

```python
mean = df["RestingBP"].mean()
std = df["RestingBP"].std()
lowest_bound = mean-3*std
highest_bound = mean+3*std
df[(df["RestingBP"]<lowest_bound)|(df["RestingBP"]>highest_bound)]
```

##the above are the outliers of RestingBP(used a z-distribution bcoz it is a normal distribution)

```python
plt.subplot(1, 2, 1)
sns.boxplot(df['Cholesterol'])
plt.title('Boxplot')

plt.subplot(1, 2, 2)
sns.distplot(df['Cholesterol'])
plt.title('Distribution Plot')


# Adjust layout and display plots
plt.tight_layout()
plt.show()
```
